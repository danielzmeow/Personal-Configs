```python
def calculate_final_score(
    lab_percent,
    midterm1_percent,
    midterm2_percent,
    final_exam_percent,
    extra_credit_points
):
    TOTAL_POINTS = 300
    WEIGHT = 0.33

    # Convert lab percentage to points
    lab_points = (lab_percent / 100) * (TOTAL_POINTS * WEIGHT)

    # Midterms: average of two midterms
    midterm_avg_percent = (midterm1_percent + midterm2_percent) / 2
    midterm_points = (midterm_avg_percent / 100) * (TOTAL_POINTS * WEIGHT)

    # Final exam
    final_exam_points = (final_exam_percent / 100) * (TOTAL_POINTS * WEIGHT)

    # Total score
    final_score = lab_points + midterm_points + final_exam_points + extra_credit_points

    return final_score


# ------------------ Example Usage ------------------

lab = 97
midterm1 = 90
midterm2 = 76
final_exam = 66
extra_credit = 20

score = calculate_final_score(
    lab,
    midterm1,
    midterm2,
    final_exam,
    extra_credit
)

print(f"Final course score: {score:.2f} / 300")
print(f"Final percentage: {(score / 300) * 100:.2f}%")
```
